```python
import pandas as pd
import numpy as np
import ptvsd
```

```python
ptvsd.enable_attach()
```

```python
nA = 4
nR = 500
df = pd.DataFrame(index=range(nR), columns=['A'+str(i) for i in range(nA)] + ['V'], 
                 data=np.random.randint(0, 10, size=[nR, nA+1]))
```

```python
group_by_cols = ['A1', 'A2', 'A3']
#gr_df = df.groupby(group_by_cols)
```

```python
def agg(par_attrs: list, df_par: pd.DataFrame):
    print('at ' + str(par_attrs))
    par_attrs_len = len(par_attrs)
    # if at the bottom of the tree
    if par_attrs_len == len(group_by_cols):
        return df_par['V'].sum()
    
    # if not, 
    gr_df_par = df_par.groupby(group_by_cols[:(par_attrs_len+1)])
    b_top = par_attrs_len == 0
    kw = {}
    for sg_attrs in gr_df_par.groups:
        df_sg = gr_df_par.get_group(sg_attrs)
        sg_attrs_key = (sg_attrs,) if b_top else sg_attrs # note: to make a tuple, add ','
        ag = agg(sg_attrs_key, df_sg)
        kw[sg_attrs_key] = ag
    
    v = np.array([kw[k] for k in kw])
    ag_p = np.sqrt(np.sum(v**2))
    return ag_p
```

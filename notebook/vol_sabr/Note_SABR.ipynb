```python
from pysabr import Hagan2002LognormalSABR
from pysabr import black
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=0.02
shift=1
t=10
v_atm_n=0.0040
beta=1
rho=0.
volvol=0.3
k=0.01
```

```python
v_x = np.linspace(-0.01, 0.05, 100)


plt.figure(figsize=(10,5))

for y_i in [-0.99,-0.5,0.0,0.5,0.99]:
    a_bs = []
    for x_i in v_x:
        sabr = Hagan2002LognormalSABR(f=f,shift=shift,t=t,v_atm_n=v_atm_n,beta=beta,rho=y_i,volvol=volvol)
        vol = sabr.normal_vol(x_i)
        a_bs.append(black.normal_call(x_i, f, t, vol, 0.0, 'call'))
    v_bs = np.array(a_bs)

    plt.plot(v_x, v_bs)

plt.show()
```

```python
v_x = np.linspace(-0.01, 0.05, 100)


plt.figure(figsize=(10,5))

for y_i in [0.0,0.5,1]:
    a_bs = []
    for x_i in v_x:
        sabr = Hagan2002LognormalSABR(f=f,shift=shift,t=t,v_atm_n=v_atm_n,beta=beta,rho=rho,volvol=y_i)
        #a_bs.append(sabr.normal_vol(x_i))
        vol = sabr.normal_vol(x_i)
        a_bs.append(black.normal_call(x_i, f, t, vol, 0.0, 'call'))
    v_bs = np.array(a_bs)

    plt.plot(v_x, v_bs)

plt.show()
```

```python
v_x = np.linspace(0,3,100)
a_bs = []
for x_i in v_x:
    sabr = Hagan2002LognormalSABR(f=f,shift=shift,t=t,v_atm_n=v_atm_n,beta=beta,rho=rho,volvol=x_i)
    a_bs.append(sabr.normal_vol(k))

v_bs = np.array(a_bs)

plt.figure(figsize=(10,5))
plt.plot(v_x, v_bs)
plt.show()
```
